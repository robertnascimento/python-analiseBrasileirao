```python
import pandas as pd
import numpy as np
```

```python
br = pd.read_csv('df_classificacao.csv')
rdd = pd.read_csv('df_rodadas.csv')
```

```python
br2019 = br[br['Temporada'] == 2019]
rdd2019 = rdd[rdd['Temporada'] == 2019]

valores_nulos = br2019.isnull().sum()
print(valores_nulos)
```

```python
rodada_38 = br2019[(br2019['Temporada'] == 2019) & (br2019['Rodada'] == 38)]
```

```python
# Somar os gols dos mandantes e visitantes para a temporada de 2019
soma_gols_mandante_2019 = rdd2019['Mandante Placar'].sum()
soma_gols_visitante_2019 = rdd2019['Visitante Placar'].sum()
soma_gols_total_br = soma_gols_mandante_2019 + soma_gols_visitante_2019

print(f"Total de gols dos mandantes na temporada de 2019: {soma_gols_mandante_2019}")
print(f"Total de gols dos visitantes na temporada de 2019: {soma_gols_visitante_2019}")
print(f"Total de gols do brasileirão na temporada de 2019: {soma_gols_total_br}")
```

```python
# Criar uma cópia do DataFrame para evitar o aviso
rdd2019 = rdd2019.copy()

# Calcular a mediana de todas as rodadas em rdd2019
mediana_total_rodadas = rdd2019['Total Gols'].median()

# Criar uma coluna com a soma total de gols por rodada usando .loc
rdd2019.loc[:, 'Total Gols'] = rdd2019['Mandante Placar'] + rdd2019['Visitante Placar']

# Calcular a mediana dos gols marcados por rodada em rdd2019
mediana_gols_por_rodada = rdd2019.groupby('Rodada')['Total Gols'].median()

# Calcular a média de gols para mandantes e visitantes em todas as rodadas de rdd2019
media_gols_mandante = rdd2019['Mandante Placar'].mean()
media_gols_visitante = rdd2019['Visitante Placar'].mean()
media_gols_rodada = media_gols_mandante + media_gols_visitante

media_formatada_mandante = "{:.2f}".format(media_gols_mandante)
media_formatada_visitante = "{:.2f}".format(media_gols_visitante)
media_gols_rodada_formatada = "{:.2f}".format(media_gols_rodada)

print(f"A mediana dos gols marcados por rodada em 2019 é:\n{mediana_gols_por_rodada}")
print(f"A média de gols para mandantes em todas as rodadas de 2019 é: {media_formatada_mandante}")
print(f"A média de gols para visitantes em todas as rodadas de 2019 é: {media_formatada_visitante}")
print(f"A média de gols em todas as rodadas de 2019 é: {media_gols_rodada_formatada}")
print(f"A mediana de todos os gols marcados em todas as rodadas de 2019 é: {mediana_total_rodadas}")

```

```python
# Criar uma cópia do DataFrame para evitar o aviso
rdd2019 = rdd2019.copy()

# Criar uma coluna com a soma total de gols por partida usando .loc
rdd2019.loc[:, 'Total Gols Partida'] = rdd2019['Mandante Placar'] + rdd2019['Visitante Placar']

# Encontrar a linha com o máximo número de gols
jogo_mais_gols = rdd2019.loc[rdd2019['Total Gols Partida'].idxmax()]

print("Jogo com mais gols:")
print(jogo_mais_gols[['Rodada','Mandante', 'Mandante Placar', 'Visitante', 'Visitante Placar', 'Total Gols Partida']])
```

```python
# Filtrar os jogos que terminaram em empate
jogos_empate = rdd2019[rdd2019['Mandante Placar'] == rdd2019['Visitante Placar']]

# Filtrar os jogos que terminaram em empate e foram 0x0
jogos_empate_0x0 = jogos_empate[jogos_empate['Mandante Placar'] == 0]

# Contar quantos jogos terminaram em empate
total_jogos_empate = len(jogos_empate)
total_jogos_empate_0x0 = len(jogos_empate_0x0)

print(f"Total de jogos que terminaram em empate: {total_jogos_empate}")
print(f"Total de jogos que terminaram em empate e foram 0x0: {total_jogos_empate_0x0}")
```

```python
# Calcular a diferença de gols por partida
rdd2019['Diferenca Gols'] = abs(rdd2019['Mandante Placar'] - rdd2019['Visitante Placar'])

# Calcular a mediana das diferenças de gols
mediana_diferenca_gols = rdd2019['Diferenca Gols'].median()

# Calcular o dobro da mediana
dobro_mediana = mediana_total_rodadas * 2

# Filtrar os jogos onde a diferença de gols é maior ou igual ao dobro da mediana
jogos_vitoria_dobro_mediana = rdd2019[(rdd2019['Diferenca Gols'] >= dobro_mediana) & (rdd2019['Mandante Placar'] > rdd2019['Visitante Placar']) | (rdd2019['Diferenca Gols'] >= dobro_mediana) & (rdd2019['Mandante Placar'] < rdd2019['Visitante Placar'])]

print("Jogos em que o time venceu por uma vantagem que é o dobro da mediana das diferenças de gols:")
print(jogos_vitoria_dobro_mediana[['Mandante', 'Visitante', 'Mandante Placar', 'Visitante Placar', 'Diferenca Gols']])
```

```python
# Time que tomou mais gols
time_mais_gols_sofridos = rodada_38[rodada_38['Gols_contra'] == rodada_38['Gols_contra'].max()]['Time'].values[0]
total_gols_sofridos = rodada_38['Gols_contra'].max()

# Time que marcou mais gols
time_mais_gols_marcados = rodada_38[rodada_38['Gols_pro'] == rodada_38['Gols_pro'].max()]['Time'].values[0]
total_gols_marcados = rodada_38['Gols_pro'].max()

# Time que sofreu menos gols
time_menos_gols_sofridos = rodada_38[rodada_38['Gols_contra'] == rodada_38['Gols_contra'].min()]['Time'].values[0]
total_gols_menos_sofridos = rodada_38['Gols_contra'].min()

# Time com mais vitórias
time_mais_vitorias = rodada_38[rodada_38['Vitoria'] == rodada_38['Vitoria'].max()]['Time'].values[0]
total_vitorias = rodada_38['Vitoria'].max()

# Time com mais derrotas
time_mais_derrotas = rodada_38[rodada_38['Derrota'] == rodada_38['Derrota'].max()]['Time'].values[0]
total_derrotas = rodada_38['Derrota'].max()

# Time com mais empates
time_mais_empates = rodada_38[rodada_38['Empate'] == rodada_38['Empate'].max()]['Time'].values[0]
total_empates = rodada_38['Empate'].max()

# Encontrar o time com mais pontos
time_mais_pontos = rodada_38[rodada_38['Pontos'] == rodada_38['Pontos'].max()]['Time'].values[0]
total_pontos = rodada_38['Pontos'].max()

print(f"Time com mais pontos: {time_mais_pontos} com um total de {total_pontos} pontos.")
print(f"Time que tomou mais gols: {time_mais_gols_sofridos} com um total de {total_gols_sofridos} gols sofridos.")
print(f"Time que marcou mais gols: {time_mais_gols_marcados} com um total de {total_gols_marcados} gols marcados.")
print(f"Time que sofreu menos gols: {time_menos_gols_sofridos} com um total de {total_gols_menos_sofridos} gols sofridos.")
print(f"Time com mais vitórias: {time_mais_vitorias} com um total de {total_vitorias} vitórias.")
print(f"Time com mais derrotas: {time_mais_derrotas} com um total de {total_derrotas} derrotas.")
print(f"Time com mais empates: {time_mais_empates} com um total de {total_empates} empates.")
```
